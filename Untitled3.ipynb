```python
import matplotlib.pyplot as plt
%matplotlib inline
import math
x = []
y1 = []
y2 = []
p = 4
for i in range (1000):
  x_ = i/1000
  x.append(x_)
  y1.append(math.sqrt(2*p*x_))
  y2.append(-math.sqrt(2*p*x_))
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.xlabel('y')
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import math
x = []
x2 = []
y = []
y2 = []
for i in range(2000):
    r = 1000
    x.append(-i)
    x2.append(i)
    y.append(np.sqrt(r ** 2 - i ** 2))
    y2.append(-np.sqrt(r ** 2 - i ** 2))
plt.figure(figsize = (6, 6))    
plt.plot(x, y, color='y')
plt.plot(x, y2, color='y')
plt.plot(x2, y2, color='y')
plt.plot(x2, y, color='y')
plt.axis('scaled')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = []
y1 = []
y2 = []
a = 3
b = 2
for i in range(-300, 301):
    xx = i/100
    x.append(xx)
    y1.append(np.sqrt((1-xx**2/a**2)*b**2))
    y2.append(-np.sqrt((1-xx**2/a**2)*b**2))
    
plt.axis([-3.5, 3.5, -3.5, 3.5])
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
y = []
x1 = []
x2 = []
a = 1
b = 1
for i in range(-300, 301):
    yy = i/100
    #if xx <= -a or xx >= a:
    y.append(yy)
    x1.append(np.sqrt((1+yy**2/b**2)/a**2))
    x2.append(-np.sqrt((1+yy**2/b**2)/a**2))
    
plt.axis([-3.5, 3.5, -3.5, 3.5])
plt.plot(x1, y)
plt.plot(x2, y)
plt.xlabel('X')
plt.ylabel('Y')
```
